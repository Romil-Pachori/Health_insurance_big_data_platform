# Claims KPI and Reimbursement Trends
Loads curated tables from PostgreSQL and computes monthly KPI trends.

```python
import os
import pandas as pd
from sqlalchemy import create_engine

host = os.getenv('POSTGRES_HOST', 'localhost')
port = os.getenv('POSTGRES_PORT', '5432')
db = os.getenv('POSTGRES_DB', 'health_insurance')
user = os.getenv('POSTGRES_USER', 'health_user')
pwd = os.getenv('POSTGRES_PASSWORD', 'health_password')
engine = create_engine(f'postgresql+psycopg2://{user}:{pwd}@{host}:{port}/{db}')
```

```python
query = '''
SELECT d.year, d.month,
       SUM(f.total_claims) AS claims_volume,
       SUM(f.total_billed) AS billed_total,
       SUM(f.total_reimbursed) AS reimbursed_total
FROM curated.fact_reimbursements f
JOIN curated.dim_date d ON f.date_sk = d.date_sk
GROUP BY d.year, d.month
ORDER BY d.year, d.month
'''
kpi_df = pd.read_sql(query, engine)
kpi_df['reimbursement_rate'] = kpi_df['reimbursed_total'] / kpi_df['billed_total']
kpi_df
```

```python
ax = kpi_df.plot(x=['year', 'month'], y=['claims_volume', 'billed_total', 'reimbursed_total'], figsize=(10, 4), marker='o')
ax.set_title('Claims Volume and Amount Trends')
```
