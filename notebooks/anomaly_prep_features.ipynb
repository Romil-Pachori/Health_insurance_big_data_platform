# Anomaly-Prep Features
Builds candidate features for suspicious provider and duplicate claim exploration.

```python
import os
import pandas as pd
from sqlalchemy import create_engine

host = os.getenv('POSTGRES_HOST', 'localhost')
port = os.getenv('POSTGRES_PORT', '5432')
db = os.getenv('POSTGRES_DB', 'health_insurance')
user = os.getenv('POSTGRES_USER', 'health_user')
pwd = os.getenv('POSTGRES_PASSWORD', 'health_password')
engine = create_engine(f'postgresql+psycopg2://{user}:{pwd}@{host}:{port}/{db}')
```

```python
query = '''
SELECT f.claim_id, f.patient_sk, f.provider_sk, f.treatment_sk, f.date_sk,
       f.billed_amount, f.reimbursed_amount,
       CASE WHEN f.billed_amount = 0 THEN 0 ELSE f.reimbursed_amount / f.billed_amount END AS reimbursement_ratio
FROM curated.fact_claims f
'''
claims = pd.read_sql(query, engine)
claims['is_high_ratio'] = (claims['reimbursement_ratio'] > 0.95).astype(int)
claims['is_potential_duplicate'] = claims.duplicated(subset=['patient_sk', 'provider_sk', 'treatment_sk', 'date_sk'], keep=False).astype(int)
claims.head()
```

```python
provider_features = claims.groupby('provider_sk', as_index=False).agg(
    claims_count=('claim_id', 'count'),
    avg_reimbursement_ratio=('reimbursement_ratio', 'mean'),
    high_ratio_count=('is_high_ratio', 'sum'),
    duplicate_pattern_count=('is_potential_duplicate', 'sum')
)
provider_features.sort_values(['high_ratio_count', 'duplicate_pattern_count'], ascending=False).head(20)
```
